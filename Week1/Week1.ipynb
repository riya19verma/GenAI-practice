Write Python code to accept user input and dynamically generate prompts.

```python
# Dynamic Prompt Generator

def generate_prompt():
    user_name = input("Enter your name: ").strip()
    topic = input("Enter a topic: ").strip()
    tone = input("Enter tone (formal / informal / creative): ").strip()

    if not user_name or not topic or not tone:
        print("Error: Input cannot be empty.")
        return

    prompt = f"""
You are an AI assistant.
Generate a {tone} explanation on the topic "{topic}"
for a user named {user_name}.
"""

    print("\nGenerated Prompt:")
    print(prompt)

# Run the function
generate_prompt()
```

Extend the pipeline to generate keywords from the summarized text.

```python
from google import genai

# ---------- CONFIGURATION ----------
client = genai.Client(api_key="API_KEY")

# ---------- SUMMARIZATION ----------
def summarize_text(text):
    if not text or not text.strip():
        raise ValueError("Input text cannot be empty")

    prompt = f"""
Summarize the following text in a clear and concise manner:

{text}
"""

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    return prompt, response.text


# ---------- KEYWORD GENERATION ----------
def generate_keywords(summary):
    if not summary or not summary.strip():
        raise ValueError("Summary cannot be empty")

    prompt = f"""
Extract 5 to 7 important keywords from the following summary.
Return only the keywords as a comma-separated list.

Summary:
{summary}
"""

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    # Convert comma-separated string into Python list
    keywords = [kw.strip() for kw in response.text.split(",")]

    return prompt, keywords


# ---------- PIPELINE EXECUTION ----------
user_text = input("Enter text: ")

try:
    summary_prompt, summary = summarize_text(user_text)
    keyword_prompt, keywords = generate_keywords(summary)

    print("\n--- Summarization Prompt ---")
    print(summary_prompt)

    print("\n--- Generated Summary ---")
    print(summary)

    print("\n--- Keyword Generation Prompt ---")
    print(keyword_prompt)

    print("\n--- Extracted Keywords ---")
    print(keywords)

except Exception as e:
    print("Error:", e)
```

Save results in both text and JSON format.

```python
from google import genai
import logging
import json

# ---------- LOGGING CONFIGURATION ----------
logging.basicConfig(
    filename="genai_pipeline.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

# ---------- GEMINI CONFIGURATION ----------
client = genai.Client(api_key="API_KEY")

# ---------- SUMMARIZATION ----------
def summarize_text(text):
    if not text or not text.strip():
        logging.error("Empty input for summarization")
        raise ValueError("Input text cannot be empty")

    prompt = f"""
Summarize the following text in a clear and concise manner:

{text}
"""
    logging.info("Summarization Prompt:\n%s", prompt)

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    summary = response.text
    logging.info("Generated Summary:\n%s", summary)

    return summary, prompt


# ---------- KEYWORD GENERATION ----------
def generate_keywords(summary):
    if not summary or not summary.strip():
        logging.error("Empty summary for keyword generation")
        raise ValueError("Summary cannot be empty")

    prompt = f"""
Extract 5 to 7 important keywords from the following summary.
Return only the keywords as a comma-separated list.

Summary:
{summary}
"""
    logging.info("Keyword Prompt:\n%s", prompt)

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    keywords = [kw.strip() for kw in response.text.split(",")]
    logging.info("Generated Keywords: %s", keywords)

    return keywords, prompt


# ---------- SAVE RESULTS ----------
def save_results(text, summary, keywords):
    # Save as TEXT file
    with open("results.txt", "w", encoding="utf-8") as f:
        f.write("Original Text:\n")
        f.write(text + "\n\n")
        f.write("Summary:\n")
        f.write(summary + "\n\n")
        f.write("Keywords:\n")
        f.write(", ".join(keywords))

    # Save as JSON file
    data = {
        "original_text": text,
        "summary": summary,
        "keywords": keywords
    }

    with open("results.json", "w", encoding="utf-8") as f:
        json.dump(data, f, indent=4)

    logging.info("Results saved to text and JSON files")


# ---------- PIPELINE EXECUTION ----------
user_text = input("Enter text: ")

try:
    summary, _ = summarize_text(user_text)
    keywords, _ = generate_keywords(summary)

    save_results(user_text, summary, keywords)

    print("\nSummary:")
    print(summary)

    print("\nKeywords:")
    print(keywords)

    print("\nResults saved successfully.")

except Exception as e:
    logging.exception("Pipeline failed")
    print("Error:", e)
```

Evaluate the generated output using custom criteria (length, relevance, clarity).

```python
# ---------- OUTPUT EVALUATION ----------
def evaluate_output(original_text, summary):
    evaluation = {}

    # 1. Length Evaluation
    original_length = len(original_text.split())
    summary_length = len(summary.split())

    if summary_length < original_length:
        evaluation["length"] = "Pass"
    else:
        evaluation["length"] = "Fail"

    # 2. Relevance Evaluation
    original_keywords = set(original_text.lower().split())
    summary_keywords = set(summary.lower().split())

    common_words = original_keywords.intersection(summary_keywords)

    if len(common_words) >= 3:
        evaluation["relevance"] = "Pass"
    else:
        evaluation["relevance"] = "Fail"

    # 3. Clarity Evaluation
    if "." in summary and len(summary.split()) > 10:
        evaluation["clarity"] = "Pass"
    else:
        evaluation["clarity"] = "Fail"

    return evaluation
```
